```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
# kde remove the line
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
# pairplot works on numerical data, it cannot work for string value 
sns.pairplot(tips)
```

```python
# hue makes the pairplot works for string and it will also add color
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='smoker')
```

```python
# the barplot is  always a graph of string against numerical values
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
# it helps calculate the standard deviation using estimator and returning th output in form of a graph
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
# it count the number of male and female that partronize the resturant
sns.countplot(x='sex',data=tips)
```

```python
#it count the number of smoker, it is mostly use to count the number of something
sns.countplot(x='smoker',data=tips)
```

```python
# it shows the graph of the bill spent and day with the highest bill, it is always a graph of string values againt numerical values
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
#it shows the concentrated graph
sns.barplot(x='day',y='total_bill', data=tips)
```

```python
sns.barplot(x='day',y='total_bill', data=tips,hue='sex')
```

```python
sns.boxplot(x='day',y='total_bill', data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='total_bill', data=tips,hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill', data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips,hue='sex')
```
